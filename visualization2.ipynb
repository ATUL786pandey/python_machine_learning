```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("E:\\sukamal sir machine learning\\Machine-learning-master\\Automobile_data.csv")
```

```python
df.head()
```

```python
# example  of seaborn
x=np.arange(10)
y=np.sqrt(x)
plt.figure(figsize=(6,6))
sns.lineplot(x,y)
plt.show()
```

```python
plt.figure(figsize=(6,6))#for doing the histogram with seaborn use displot
sns.distplot(df["width"])
plt.show()
```

```python
width_mean=df["width"].mean()
width_median=df["width"].median()
plt.figure(figsize=(6,6))
sns.distplot(df["width"])
plt.axvline(width_mean,color="red")
plt.axvline(width_median,color="green")
plt.show()
```

```python
#categorial data in seaborn we used to do count plot 
# count plot
plt.figure(figsize=(8,8))
sns.countplot(data=df,x="body-style")
plt.show()
```

```python
#count plot for make
plt.figure(figsize=(8,8))
sns.countplot(data=df,y="make")
plt.show()
```

```python
df["fuel-type"].value_counts()
```

```python
#categorial data in seaborn we used to do count plot 
# count plot
plt.figure(figsize=(8,8))
sns.countplot(data=df,x="body-style",hue="fuel-type")
plt.show()
```

# heatmap 

cor=df.corr()

```python
sns.heatmap(cor,annot=True)
plt.show()
#light color show positive cohrelation and dark color show negative corelation 
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(cor,annot=True,cmap="viridis")
plt.show()
```

# scatter plot in seaborn

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=df,x="width",y="price")
plt.plot()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(data=df,x="width",y="price",hue="body-style")
plt.plot()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=df,x="width",y="price",hue="make")
plt.plot()
```

# box plot

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x="price")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(data=df,x="price",y="make")
plt.show()
```

# sub plot

```python
#  to do the subplot just put fig,axes=plt.subplot(2,2,figsize=()) and after that put all the hist  graph.....etc
x=[1,2,3,4]
y=[3,4,5,6]

fig,axes=plt.subplots(2,2,figsize=(8,6))
axes[0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes[1,0].plot(x,y)
axes[1,1].scatter(x,y)
plt.show()
```

```python
fig,axes=plt.subplots(2,2,figsize=(15,14))
sns.distplot(df["width"],ax=axes[0,0])
sns.countplot(data=df,x="body-style",hue="fuel-type",ax=axes[0,1])
sns.scatterplot(data=df,x="width",y="price",ax=axes[1,0])
sns.boxplot(data=df,x="price",y="make",ax=axes[1,1])
plt.show()
```
