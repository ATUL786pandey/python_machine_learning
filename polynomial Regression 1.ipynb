```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
```

```python
x=np.array([1,3,7,9,11])
y=np.array([11,12,13,25,30])
```

# Gradient Desent

```python
def gradient_desent(x,y):
    n=len(x)
    learning_rate=0.001
    intercept=slope=0
    iteration=500
    
    for i in range(iteration):
        y_pred=x*slope+intercept
        mse=mean_squared_error(y_pred,y)
        
        md=-(2/n)*sum(x*(y-y_pred))
        bd=-(2/n)*sum(y-y_pred)
        
        slope=slope-(learning_rate * md)
        intercept=intercept-(learning_rate*bd)
        
        print("slope: ",slope,"intercept :",intercept,"mse :",mse,"iteration:",i)
```

```python
gradient_desent(x,y)
```

#SKlearn Implemetation

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
sg=SGDRegressor()
```

```python
sg.fit(x_train.reshape((-1,1)),y_train)
```

```python
y_pred=sg.predict(x_test.reshape((-1,1)))
```

```python
mean_squared_error(y_pred,y_test)
```

# polynomial Regressor

```python
#Linear
def poly1(x):
    y=4*x+5
    return y

#Quadratic

def ploy_2(x):
    y=6 * (x**2) + 4*x + 5
    return y

# Cubic

def ploy_3(x):
    y=8 *(x**3)+6 * (x**2)+4*x+5
    return y

#plot
def plot(x,y):
    plt.figure()
    plt.plot(x,y)
    plt.show()
```

```python
x=np.arange(-100,100)
```

```python
# Plot Linear (for single degree)

y=poly1(x)
plot(x,y)
```

```python
#plot for quadratic (for 2 degree)
y=ploy_2(x)
plot(x,y)
```

```python
#plot for cubic (for 3 degree)
y=ploy_3(x)
plot(x,y)
```
