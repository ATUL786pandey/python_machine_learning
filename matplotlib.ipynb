```python
area=[600,800,1000,1200,1500]
price=[15,20,25,35,40]
plt.plot(area,price)
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(area,price)
plt.title("area vs price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(area,price)
plt.title("area vs price")
plt.xlabel("Area")
plt.ylabel("Price")
plt.xlim(200,2000)
plt.ylim(10,50)
plt.grid()
plt.show()
```

```python
x=np.arange(10,50)
y1=np.sqrt(x)
y2=np.log(x)
plt.figure(figsize=(5,7))
plt.plot(x,y1,color="orange")
plt.plot(x,y2,color="red")
plt.legend(labels=["square root","Logs"])

plt.grid()
plt.show()
#square and log are used to reduce skewnesss
```

```python
x=np.arange(10,50)
y1=np.sqrt(x)
y2=np.log(x)
plt.figure(figsize=(5,7))
plt.plot(x,y1,color="orange")
plt.plot(x,y2,color="red")
plt.legend(labels=["square root","Logs"])
plt.xticks(rotation=45)
plt.yticks(rotation=90)

plt.grid()
plt.show()
#square and log are used to reduce skewnesss
```
